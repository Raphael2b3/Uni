```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-100, 100, 3000)
y = x

X,Y = np.meshgrid(x,y)

```

```python
f = 5
R1 = np.sqrt(X**2 + (Y-1/f)**2)
R2 = np.sqrt(X**2 + (Y-1/(2*f))**2)
R3 = np.sqrt(X**2 + Y**2)

phi = np.pi/2/3e8
sender1 = np.exp(1j*2*np.pi*f*R1)*np.exp(-4j*phi)
sender2 = np.exp(1j*2*np.pi*f*R2)*np.exp(-3j*phi)
sender3 = np.exp(1j*2*np.pi*f*R3)*np.exp(-2j*phi)

plt.imshow(np.abs(sender1+sender2+sender3), extent=(-100, 100, -100, 100),cmap="viridis")
plt.scatter([50],[25], color="red")
```
