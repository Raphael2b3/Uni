# Jupyter Notebook
In Jupyter-Notebooks bestehen Dokumente aus **Zellen**, die es ermöglichen, Inhalte in einem flexiblen Format anzuordnen. Es gibt zwei Haupttypen von Zellen:

1. **Code-Zellen**: In diesen Zellen können Programmcode (z. B. in Python) geschrieben und ausgeführt werden. Das Ergebnis der Code-Ausführung wird direkt unter der Zelle angezeigt, sobald sie ausgeführt wird.
   
   Beispiel:
   ```python
   print("Hallo, Jupyter!")

````python
print("Hello World!")
````

2. **Markdown-Zellen**: Diese Zellen werden verwendet, um formatierten Text, Bilder, Links, mathematische Ausdrücke (LaTeX), etc. hinzuzufügen. Sie sind hilfreich für Dokumentation und Erklärungen in einem Notebook.

Beispiel:

```markdown
# Hello

World!
```

# Kernel

Der Kernel ist der Rechenkern eines Jupyter-Notebooks. Es handelt sich um einen Hintergrundprozess, der für die Ausführung des Codes verantwortlich ist, der in den Zellen des Notebooks steht. Jede Programmiersprache, die von Jupyter unterstützt wird (wie Python, R, Julia), benötigt einen entsprechenden Kernel. Im Fall von Python ist der Kernel in der Regel ein IPython-Kernel.
